# Regression

### - Linear Regression
### - SVM
### - Stochastic Gradient Descent
### - Lasso
### - Ridge
#### We can add more if you want :)

```python
Y = df['GPA'].values
x = df1.values

Xtrain, Xtest, ytrain, ytest = train_test_split(x, Y, test_size=0.3)

classifiers = [('LinearRegression', LinearRegression()),
               ('SVM', svm.SVR()),
               ('Stochastic Gradient Descent', SGDRegressor(max_iter=180, tol=1e-3)),
               ('Lasso', Lasso(alpha=0.01)),
               ('Ridge', Ridge(alpha=0.01))
              ]
allscores_reg = []

for name, classifier in classifiers:
    scores_reg = []
    for i in range(30):
        roc = cross_val_score(classifier, x, Y)
        scores_reg.extend(list(roc))
    scores_reg = np.array(scores_reg)
    print(name, scores_reg.mean())
    new_data = [(name, score) for score in scores_reg]
    allscores_reg.extend(new_data)
```

# Now lets visualize scores 

```python
temp = pd.DataFrame(allscores_reg, columns=['classifier', 'score'])
plt.figure(figsize=(15,10))
sns.factorplot(x='classifier', 
               y="score",
               data=temp, 
               saturation=1, 
               kind="box", 
               ci=None, 
               aspect=1, 
               linewidth=1, 
               size = 10)     
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
```

# Data Encoding For Classification

```python
df2 = pd.DataFrame.copy(df1)

df2 = df2.assign(gpaa=pd.Series())
df2 = df2.assign(cgpaa=pd.Series())

df2.loc[df2.GPA == 0, ["gpaa"] ] = 0
df2.loc[(df2.GPA >= 1) & (df2.GPA < 1.67), ["gpaa"] ] = 1
df2.loc[(df2.GPA >= 1.67) & (df2.GPA < 2.67), ["gpaa"] ] = 2
df2.loc[(df2.GPA >= 2.67) & (df2.GPA < 3.67), ["gpaa"] ] = 3
df2.loc[(df2.GPA >= 3.67) & (df2.GPA <= 4), ["gpaa"] ] = 4

df2.loc[df2.CGPA == 0, ["cgpaa"] ] = 0
df2.loc[(df2.CGPA >= 1) & (df2.CGPA < 1.67), ["cgpaa"] ] = 1
df2.loc[(df2.CGPA >= 1.67) & (df2.CGPA < 2.67), ["cgpaa"] ] = 2
df2.loc[(df2.CGPA >= 2.67) & (df2.CGPA < 3.67), ["cgpaa"] ] = 3
df2.loc[(df2.CGPA >= 3.67) & (df2.CGPA <= 4), ["cgpaa"] ] = 4

df2 = df2.drop(['GPA','CGPA'],axis=1)
df2.rename(columns = {'gpaa':'GPA'}, inplace = True)
df2.rename(columns = {'cgpaa':'CGPA'}, inplace = True)


df2
```

# Classification

```python
Y = df2['GPA']
x = df2.values
Xtrain, Xtest, ytrain, ytest = train_test_split(x, Y, test_size=0.3)

classifiers = [('RandomForestClassifierG', RandomForestClassifier(n_jobs=-1, criterion='entropy')),
               ('RandomForestClassifierE', RandomForestClassifier(n_jobs=-1, criterion='entropy')),
               ('AdaBoostClassifier', AdaBoostClassifier()),
               ('ExtraTreesClassifier', ExtraTreesClassifier(n_jobs=-1)),
               ('KNeighborsClassifier', KNeighborsClassifier(n_jobs=-1)),
               ('DecisionTreeClassifier', DecisionTreeClassifier()),
               ('ExtraTreeClassifier', ExtraTreeClassifier()),
               ('LogisticRegression', LogisticRegression()),
               ('GaussianNB', GaussianNB()),
               ('BernoulliNB', BernoulliNB())
              ]
allscores_class = []

for name, classifier in classifiers:
    scores_class = []
    for i in range(20): 
        roc = cross_val_score(classifier, x, Y)
        scores_class.extend(list(roc))
    scores_class = np.array(scores_class)
    print(name, scores_class.mean())
    new_data = [(name, score) for score in scores_class]
    allscores_class.extend(new_data)
```

# Lets visualize scores

```python
temp = pd.DataFrame(allscores_class, columns=['classifier', 'score'])
plt.figure(figsize=(15,10))
sns.factorplot(x='classifier', 
               y="score",
               data=temp, 
               saturation=1, 
               kind="box", 
               ci=None, 
               aspect=1, 
               linewidth=1, 
               size = 10)     
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
```

## Exploring features that affected a certain classifier

```python
radm = RandomForestClassifier()
radm.fit(Xtrain, ytrain)

clf = radm
indices = np.argsort(radm.feature_importances_)[::-1]

# Print the feature ranking
print('Feature ranking:')

for f in range(df1.shape[1]):
    print('%d. feature %d %s (%f)' % (f+1 , 
                                      indices[f], 
                                      df1.columns[indices[f]], 
                                      radm.feature_importances_[indices[f]]))
```
